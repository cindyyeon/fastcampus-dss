### 6주차 scipy로 공부하는 확률분포

```python
import scipy as sp
import scipy.stats
```

```python
rv = sp.stats.norm(loc = 1, scale = 2)
```

```python
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title('probabilty density function')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

#### 8.5 스튜던트 t분포, 카이제곱분포, F분포

```python
# 연습문제 8.5.1
import scipy as sp
rv = sp.stats.norm()
M = 2000

plt.subplot(1,2,1)
N = 6
x = rv.rvs((N, M))
t = (x ** 2).sum(axis = 0)
sns.distplot(t, kde = False)
plt.show()

plt.subplot(1,2,2)
N = 30
x = rv.rvs((N, M))
t = (x ** 2).sum(axis = 0)
sns.distplot(t, kde = False)
plt.show()
```
